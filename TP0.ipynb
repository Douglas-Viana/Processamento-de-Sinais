### QUESTÃO 1
Crie um script que permita a entrada da amplitude, frequência, tempo inicial, passo de tempo e tempo final pelo usuário, além de perguntar se quer ver o gráfico final (N = não, S = sim). A saída deverá repetir de forma organizada a entrada dos dados e apresentar ou não o gráfico de acordo com o que o usuário solicitou. Para teste, entre com amplitude = 1, frequência = 60 Hz e t de 0 a 0,1 s com dt = 1 ms.

Nota: o gráfico deve ter título, curva em preto com os pontos do vetor
em azul, eixo x e y definidos.

```python
import matplotlib.pyplot as plt
import numpy as np

print("---------Questão 1---------")

amp = input('Digite o valor da amplitude do sinal: ')
f = input('Digite o valor da frequência do sinal: ')
ti = input('Digite o valor de tempo inicial do sinal: ')
dt = input('Digite o valor de passo de tempo: ')
tf = input('Digite o valor de tempo final do sinal: ')

op_1 = input('Deseja ver o gráfico final: (N=não, S=sim)')

print("\n---------RESULTADOS---------")

print('AMPLITUDE DO SINAL: ' + amp)
print('FREQUÊNCIA DO SINAL: ' + f + ' Hz')
print('INTERVALO DE TEMPO: t = '+ ti + 's a ' + tf + 's - PASSO DE TEMPO: ' + dt + 's')

if op_1 == 's' or op_1 == 'S':
    amp = float(amp)
    f = float(f)
    ti = float(ti)
    dt = float(dt)
    tf = float(tf)
    
    t = np.arange(ti, tf+dt, dt)
    y = amp*np.sin(2*np.pi*f*t)
    
    plt.figure(figsize=(10,5))

    plt.plot(t, y, 'k', t, y, 'b.')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Amplitude')
    plt.title('y = sen(t)', fontsize=15)
    
    plt.grid()
    plt.show()

```

### QUESTÃO 2
Um mixer permite a um DJ utilizar até três ondas simultaneamente para compor um trecho de música. Considere as ondas a seguir:

Onda 1: 2.cos(ωt), f = 300 Hz;

Onda 2: 0,8.cos(ωt), f = 900 Hz;

Onda 3: 0,4.cos(ωt), f = 2 kHz.

Pede-se uma rotina que imprima as opções de onda disponíveis e as opções para o usuário escolher ($2^{3}$-1 opções). Após a escolha, a rotina deve indicar a opção escolhida pelo usuário e as ondas relacionadas, além de apresentar um gráfico com a onda gerada pela composição
final. Usar dt = 50 μs e tf = 10 ms.

Nota: o gráfico final deve ter título (opção escolhida), curva em preto e
eixos x e y identificados (tempo (s) e amplitude (1)).

```python
import matplotlib.pyplot as plt
import numpy as np

print('---------Questão 2---------')

print('Opções de onda disponíveis para mixagem:')
print('1 --> a')
print('2 --> b')
print('3 --> c')
print('4 --> a + b')
print('5 --> a + c')
print('6 --> b + c')
print('7 --> a + b + c')

print('\nLegenda:')
print('a = 2xcos(2x\u03C0xfxt), f = 300 Hz')
print('b = 0.8xcos(2x\u03C0xfxt), f = 900 Hz')
print('c = 0.4xcos(2x\u03C0xfxt), f = 2 kHz')

op_2 = input('Digite a opção desejada: ')

print("\n---------RESULTADOS---------")
print('OPÇÃO ESCOLHIDA PELO USUÁRIO: '+ op_2)

tf = 10e-3
dt = 50e-6
t = np.arange(0, tf+dt, dt)

if op_2 == '1':
    
    print('Opção 1 --> Onda a')
    ya = 2*np.cos(2*np.pi*300*t)
    
    plt.figure(figsize=(10,5))
    
    plt.plot(t ,ya , color='k')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Amplitude (1)')
    plt.title('Opção 1: Onda a', fontsize=15)
    plt.grid()
    plt.show()
    
elif op_2 == '2':
    
    print("Opção 2 --> Onda b")
    yb=0.8*np.cos(2*np.pi*900*t)
    
    plt.figure(figsize=(10,5))
    
    plt.plot(t, yb, color='k')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Amplitude (1)')
    plt.title('Opção 2: Onda b', fontsize=15)
    plt.grid()
    plt.show()
    
elif op_2 == '3':
    
    print("Opção 3 --> Onda c")
    yc=0.4*np.cos(2*np.pi*2e3*t)
    
    plt.figure(figsize=(10,5))
    
    plt.plot(t, yc, color='k')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Amplitude (1)')
    plt.title('Opção 3: Onda c', fontsize=15)
    plt.grid()
    plt.show()
    
elif op_2 =='4':
    
    print("Opção 4 --> Onda a + b")
    ya=2*np.cos(2*np.pi*300*t)
    yb=0.8*np.cos(2*np.pi*900*t)
    
    plt.figure(figsize=(10,5))
    
    plt.plot(t, ya+yb, color='k')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Amplitude (1)')
    plt.title('Opção 4: Onda a + b', fontsize=15)
    plt.grid()
    plt.show()
    
elif op_2 =='5':
    
    print("Opção 5 --> Onda a + c")
    ya=2*np.cos(2*np.pi*300*t)
    yc=0.4*np.cos(2*np.pi*2e3*t)
    
    plt.figure(figsize=(10,5))
    
    plt.plot(t, ya+yc, color='k')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Amplitude (1)')
    plt.title('Opção 5: Onda a + c', fontsize=15)
    plt.grid()
    plt.show()
    
elif op_2 == '6':
    
    print("Opção 6 --> Onda b + c")
    yb=0.8*np.cos(2*np.pi*900*t)
    yc=0.4*np.cos(2*np.pi*2e3*t)
    
    plt.figure(figsize=(10,5))
    
    plt.plot(t, yb+yc, color='k')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Amplitude (1)')
    plt.title('Opção 6: Onda b + c', fontsize=15)
    plt.grid()
    plt.show()
    
elif op_2 == '7':
    
    print("Opção 7 --> Onda a + b + c")
    ya=2*np.cos(2*np.pi*300*t)
    yb=0.8*np.cos(2*np.pi*900*t)
    yc=0.4*np.cos(2*np.pi*2e3*t)
    
    plt.figure(figsize=(10,5))
    
    plt.plot(t, ya+yb+yc ,'k')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Amplitude (1)')
    plt.title('Opção 7: Onda a + b + c', fontsize=15)
    plt.grid()
    plt.show()
    
else:
    print("Opção não encontrada! Retorne ao menu.")
```

### QUESTÃO 3
Funções no domínio do tempo contínuo (DTC) e discreto (DTD) devem ser comparadas (figura composta por dois subplots). Considere o sinal no DTC com frequência oito vezes maior que o DTD. O usuário deve escolher a opção desejada e será apresentado um subplot, contendo a onda DTC em linha contínua e a onda DTD na forma stem.

Sinal 1: onda senoidal, com amplitude 1 e f = 60 Hz (dt = 0,2 ms e tf = 0,1 s);

Sinal 2: onda exponencial decrescente, com amplitude 1 e constante de tempo igual a 0,1 (dt = 1 ms e tf = 0,5 s);

Sinal 3: Sinal 1 somado à onda senoidal, com amplitude 1 e f = 300 Hz (dt = 0,2 ms e tf = 0,1 s).

```python
import matplotlib.pyplot as plt
import numpy as np

print('---------Questão 3---------\n')
print('Opções de onda disponíveis para mixagem:')
print('1 --> Senoide, amplitude unitária, 60 Hz, tempo de 0 a 0.1 s, dt = 0.2 ms')
print('2 --> Exponencial decrescente, amplitude unitária, constante de tempo = 0.1 s, tempo de 0 a 0.5 s, dt = 1 ms')
print('3 --> Sinal 1 adicionado à outra senoide, amplitude unitária, 300 Hz, tempo de 0 a 0.1 s, dt = 0.2 ms')

op_3 = input('\nDigite a opção desejada: ')

print('\n---------RESULTADOS---------\n')
print('Opção escolhida pelo usuário: ' + op_3)

if op_3 == '1':
    print('1 --> Senoide, amplitude unitária, 60 Hz, tempo de 0 a 0.1 s, dt = 0.2 ms')
    
    tf = 0.1
    dt = 0.2e-3
    t1 = np.arange(0,tf+dt,dt)
    y1 = np.sin(2*np.pi*60*t1)
    
    dt2 = 8*dt
    t2 = np.arange(0,tf+dt2,dt2)
    n = np.arange(0,len(t2))
    y2 = np.sin(2*np.pi*60*t2)
    
    plt.figure(figsize=(10,5))
    
    plt.subplot(2,1,1)
    plt.plot(t1, y1, color='k')
    
    plt.subplot(2,1,2)
    plt.stem(n, y2)
    
    plt.show()
    
elif op_3 == '2':
    print('2 --> Exponencial decrescente, amplitude unitária, constante de tempo = 0.1 s, tempo de 0 a 0.5 s, dt = 1 ms')
    
    tf = 0.5
    dt = 1e-3
    tau = 0.1
    
    t1 = np.arange(0,tf+dt,dt)
    y1 = np.exp(-t1/tau)
    
    dt2 = 8*dt
    t2 = np.arange(0,tf+dt2,dt2)
    y2 = np.exp(-t2/tau)
    n = np.arange(0,len(t2))
    
    plt.figure(figsize=(10,5))
    
    plt.subplot(2,1,1)
    plt.plot(t1, y1, color='k')
    
    plt.subplot(2,1,2)
    plt.stem(n, y2)
    
    plt.show()
    
elif op_3 == '3':
    print('3 --> Sinal 1 adicionado à outra senoide, amplitude unitária, 300 Hz, tempo de 0 a 0.1 s, dt = 0.2 ms')
    
    tf1 = 0.1
    dt1 = 0.2e-3
    t1 = np.arange(0,tf1+dt1,dt1)
    y1a = np.sin(2*np.pi*60*t1)
    y2a = np.sin(2*np.pi*300*t1)
    
    dt2 = 0.2e-3*8
    t2 = np.arange(0,tf1+dt2,dt2)
    y1b = np.sin(2*np.pi*60*t2)
    y2b = np.sin(2*np.pi*300*t2)
    n = np.arange(0,len(t2))
    
    plt.figure(figsize=(10,5))
    
    plt.subplot(2,1,1)
    plt.plot(t1, y1a+y2a, color='k')
    
    plt.subplot(2,1,2)
    plt.stem(n, y1b+y2b)
    
    plt.show()
    
else:
    print("Opção não encontrada! Retorne ao menu.")
```

### QUESTÃO 4
Considerando as opções disponíveis na questão 2, crie uma rotina que imprima todas as ondas em uma única figura, mas na forma de subplots. Obs: Todos os gráficos devem ter a função grid ativada e títulos correspondentes às opções da questão 2 (Opção 1, Opção 2, ..., Opção 7).



```python
import matplotlib.pyplot as plt
import numpy as np

print('\nLegenda:')
print('a = 2xcos(2x\u03C0xfxt), f = 300 Hz')
print('b = 0.8xcos(2x\u03C0xfxt), f = 900 Hz')
print('c = 0.4xcos(2x\u03C0xfxt), f = 2 kHz')

tf = 10e-3
dt = 50e-6
t = np.arange(0, tf+dt, dt)

plt.figure(figsize=(10,5))

#Opção 1: Onda a
ya = 2*np.cos(2*np.pi*300*t)
plt.subplot(4,2,1)
plt.plot(t ,ya)
plt.title('Opção 1: Onda a', fontsize=11, fontweight='bold')
plt.grid()

#Opção 2: Onda b
yb=0.8*np.cos(2*np.pi*900*t)
plt.subplot(4,2,2)
plt.plot(t, yb)
plt.title('Opção 2: Onda b', fontsize=11, fontweight='bold')
plt.grid()

#Opção 3: Onda c 
yc=0.4*np.cos(2*np.pi*2e3*t)
plt.subplot(4,2,3)
plt.plot(t, yc)
plt.title('Opção 3: Onda c', fontsize=11, fontweight='bold')
plt.grid()

#Opção 4: Onda a + b
ya=2*np.cos(2*np.pi*300*t)
yb=0.8*np.cos(2*np.pi*900*t)
plt.subplot(4,2,4)
plt.plot(t, ya+yb)
plt.title('Opção 4: Onda a + b', fontsize=11, fontweight='bold')
plt.grid()

#Opção 5: Onda a + c
ya=2*np.cos(2*np.pi*300*t)
yc=0.4*np.cos(2*np.pi*2e3*t)
plt.subplot(4,2,5)
plt.plot(t, ya+yc)
plt.title('Opção 5: Onda a + c', fontsize=11, fontweight='bold')
plt.grid()

#Opção 6: Onda b + c
yb=0.8*np.cos(2*np.pi*900*t)
yc=0.4*np.cos(2*np.pi*2e3*t)
plt.subplot(4,2,6)
plt.plot(t, yb+yc)
plt.title('Opção 6: Onda b + c', fontsize=11, fontweight='bold')
plt.grid()

#Opção 7: Onda a + b + c
ya=2*np.cos(2*np.pi*300*t)
yb=0.8*np.cos(2*np.pi*900*t)
yc=0.4*np.cos(2*np.pi*2e3*t)
plt.subplot(4,1,4)
plt.plot(t, ya+yb+yc)
plt.title('Opção 7: Onda a + b + c', fontsize=11, fontweight='bold')
plt.grid()

plt.subplots_adjust(hspace=0.9, wspace=0.20)
plt.show()
    
```
